# 4. If statements

**If-Statements with User-Defined Functions**

```python
# Define price_label function to dataframe

def price_label(row):
    
    if row['prices'] <=5:
        return 'Low range product'
    elif (row['prices'] >5) and (row['prices'] <=15):
        return 'Mid range product'
    elif row ['prices'] >15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the "price_label" to subset dataframe (df)

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values in the new column "price_range"

df['price_range'].value_counts(dropna = False)
```

```python
# Check for the max price within the subset datafram (df)

df['prices'].max()
```

**If-Statements with the loc() Function**

```python
# Create conditions

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <=15) & (df['prices'] >5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <=5, 'price_range_loc'] = 'Low range product'
```

```python
# Check values in the new column "price_range_loc" (in the subset dataframe)

df['price_range_loc'].value_counts(dropna=False)
```

```python
# Add "price_range_loc" to the entire dataframe by using loc function

df_ords_prods.loc[df_ords_prods['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df_ords_prods.loc[(df_ords_prods['prices'] <=15) & (df_ords_prods['prices'] >5), 'price_range_loc'] = 'Mid range product'
```

```python
# Create a for-loop 

result2 = []

for value in df_ords_prods['orders_day_of_week']:
  if value == 0 or value == 1:
    result2.append("Two busiest days")
  elif value == 3 or value == 4:
    result2.append("Two slowest days")
  else:
    result2.append("Regularly busy")
```

```python
# Define busiest days to dataframe

def busiest_days(row):
    
    if (row['orders_day_of_week'] ==0) or (row['orders_day_of_week'] ==1):
        return 'Two busiest days'
    elif (row['orders_day_of_week'] ==3) or (row['orders_day_of_week'] ==4):
        return 'Two slowest days'
    elif (row ['orders_day_of_week'] ==2) or (row['orders_day_of_week'] ==5) or (row['orders_day_of_week'] ==6):
        return 'Regularly busy'
    else: return np.nan
```

# 9. 4.7:4 
When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check the current columns name

df_ords_prods.head()
```

```python
# Check frequency of column "order_hour_of_day"

df_ords_prods['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Define busiest_period_of_day to dataframe

def busiest_period_of_day(row):

  if (row['order_hour_of_day'] == 10) or (row['order_hour_of_day'] == 11) or (row['order_hour_of_day'] == 14) or (row['order_hour_of_day'] == 15) or (row['order_hour_of_day'] == 13) or (row['order_hour_of_day'] == 12) or (row['order_hour_of_day'] == 16):
    return 'Most orders between 10-16'
  elif (row['order_hour_of_day'] == 3) or (row['order_hour_of_day'] == 4) or (row['order_hour_of_day'] == 2) or (row['order_hour_of_day'] == 5) or (row['order_hour_of_day'] == 1) or (row['order_hour_of_day'] == 0) or (row['order_hour_of_day'] == 6):
    return 'Fewest orders between 00-06'
  else: return 'Average orders'
```

```python
# Apply the "busiest_period_of_day" to dataframe

df_ords_prods['busiest_period_of_day'] = df_ords_prods.apply(busiest_period_of_day, axis=1)
```
